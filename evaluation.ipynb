```python
from scipy.io import wavfile
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

sampling_rate, data = wavfile.read('./test_sine/enhanced_audio.wav')
freqs, ps = signal.welch(data, fs=sampling_rate, window='hann', nperseg = 4096, noverlap=None, nfft=None,
             return_onesided=True, scaling='density', axis=- 1, average='mean')

plt.semilogy(freqs, ps/np.max(np.abs(ps)))
# plt.ylim([0.5e-3, 1])
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD')
plt.show()
```
